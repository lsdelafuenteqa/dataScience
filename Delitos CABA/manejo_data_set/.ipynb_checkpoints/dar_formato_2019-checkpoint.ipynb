```python
import pandas as pd
```

```python
df = pd.read_csv("final-final.csv")
```

```python
df
```

```python
df["tipo"].unique()
```

```python
df.replace(to_replace="Lesiones", value = "Lesiones Seg Vial", inplace= True)
```

```python
df.replace(to_replace="Homicidio", value = "Homicidio Seg Vial", inplace= True)
```

```python
df.replace(to_replace="Robo (con violencia)", value = "Robo (Con violencia)", inplace= True)
```

```python
df.replace(to_replace="Hurto (sin violencia)", value = "Hurto (Sin violencia)", inplace= True)
```

```python
df.tipo.value_counts()
```

```python
df["cantidad_vehiculos"].replace(to_replace="Comuna 2", value = 2, inplace=True)
```

```python
df["cantidad_vehiculos"].replace(to_replace="Comuna 1", value = 1, inplace=True)
```

```python
df["cantidad_vehiculos"].fillna(0, inplace=True)
```

```python
df["cantidad_victimas"].fillna(0, inplace=True)
```

```python
df
```

```python
df["id"].dropna(inplace=True)
df["id"].isna().sum()
```

```python
df["id"] = df.id.astype(int)
df['cantidad_vehiculos'] = df.cantidad_vehiculos.astype(int)
df['cantidad_victimas'] = df.cantidad_victimas.astype(int)
```

```python
df.drop(columns="subtipo", inplace=True)
```

```python
df.rename(columns={"tipo":"tipo_delito"}, inplace=True)
```

```python
df.to_csv('deilitos2019.csv', header=False, index=False)
```
